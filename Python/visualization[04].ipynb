## 1차원 분포 플롯
- barplot() : 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 차트
- 평균 : 막대의 높이,  편차 : 에러바

```python
iris    = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips    = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
# barplot
# 스타일 테마 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

# 그래프 객체를 생성하기 위한 figure객체 얻어오기
fig = plt.figure(figsize=(15,5))


area01 = fig.add_subplot(1,3,1)
area01.set_title('titanic survived sex')
area02 = fig.add_subplot(1,3,2)
area02.set_title('titanic survived sex - hue')
area03 = fig.add_subplot(1,3,3)
area03.set_title('titanic survived sex - hue dodge')

sns.barplot(ax=area01)
sns.barplot(ax=area02)
sns.barplot(ax=area03)

#성별에 따른 생존률
sns.barplot(x='sex', y='survived', data=titanic, ax=area01)

# hue = 
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=area02)

# dodge //누적해서 그리기
sns.barplot(x='sex', y='survived', hue='class', dodge=False, data=titanic, ax=area03)

plt.show()
```

```python
## tips - barplot()
## 요일별 전체 팁을 시각화

fig=plt.figure(figsize=(15,5))

sns.barplot(x='day',y='total_bill',data=tips)

plt.title("요일별 전체 팁")
plt.show()
```

### # countplot() : 카테고리 값별로 데이터 빈도를 표시하는 차트

```python
# barplot
# 스타일 테마 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체를 생성하기 위한 figure객체 얻어오기
fig = plt.figure(figsize=(15,5))

area01 = fig.add_subplot(1,3,1)
area01.set_title('titanic class')

area02 = fig.add_subplot(1,3,2)
area02.set_title('titanic class - who')

area03 = fig.add_subplot(1,3,3)
area03.set_title('titanic class - who(stacked)')

#기본
sns.countplot(x='class', data=titanic, ax=area01)
# hue = 
sns.countplot(x='class', hue='who', palette='Set3', data=titanic, ax=area02)

# dodge //누적해서 그리기
sns.countplot(x='class', hue='who', dodge=False, palette='Set2', data=titanic, ax=area03)

plt.show()
```

```python
## tips를 이용한 countplot()
# 요일별로 팁을 준 빈도를 시각화 한다면?
fig=plt.figure(figsize=(15,5))
sns.countplot(x='day', data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

### # stripplot() : 스카터플롯처럼 모든 데이터를 점으로 시각화
### # jitter = True 설정하면 위치를 무작위로 바꿔서 겹치지 않게 해준다

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.title('요일별 전체 팁')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('요일별 전체 팁')
plt.show()
```

```python
# 타이타닉 데이터를 이용하여
# 서브플롯 2개 영역으로 만들고
# class 별 나이의 분포를 확인하라 ( stripplot(), swarmplot() )


# 그래프 객체를 생성하기 위한 figure객체 얻어오기
fig = plt.figure(figsize=(15,5))

area01 = fig.add_subplot(1,2,1)
area01.set_title('titanic class')

area02 = fig.add_subplot(1,2,2)
area02.set_title('titanic class - who')

#기본
sns.stripplot(x='class', y='age', data=titanic, ax=area01,  jitter=True)
# hue = 
sns.swarmplot(x='class', y='age', palette='Set1', data=titanic, ax=area02)

plt.show()
```

```python
# 만약에 tips에서
# 요일별, 성별 전체 팁 stripplot()시각화
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips,jitter=True )
plt.title('요일별 전체 팁')
plt.legend(loc='best')
plt.show()
```

### # boxplot(), violinplot()

```python
# 스타일 테마 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체를 생성하기 위한 figure객체 얻어오기
fig = plt.figure(figsize=(15,10))

ax01 = fig.add_subplot(2,2,1)
ax01.set_title('box')

ax02 = fig.add_subplot(2,2,2)
ax02.set_title('box')

ax03 = fig.add_subplot(2,2,3)
ax03.set_title('violin')

ax04 = fig.add_subplot(2,2,4)
ax04.set_title('violin')


sns.boxplot(x='alive', y='age', data=titanic, ax=ax01)
sns.boxplot(x='alive', y='age', hue='sex',data=titanic, ax=ax02)

sns.violinplot(x='alive', y='age', data=titanic, ax=ax03)
sns.violinplot(x='alive', y='age', hue='sex',data=titanic,ax=ax04)

plt.show()
```

### # heatmap : 하나 이상의 카테고리 값의 변화를 색 변화로 시각화하는 차트

```python
# titanic
# 피벗테이블로 범주형 변수를 각각 행, 열로 재구분하여 정리
# 인덱스 성별, 열 클래스 
table = titanic.pivot_table(index=['sex'], columns =['class'], aggfunc='size')
#table.head()

sns.heatmap(table,
           annot=True, fmt='d', linewidth=1)
plt.show()
```

```python
# flights 연도, 월 별 승객수에 대한 heatmap 만들기
# x 월, y 연도

table02 = flights.pivot_table(index=['month'], columns =['year'], aggfunc='sum')
table02.head()

sns.heatmap(table02,
           annot=True, fmt='d', linewidth=1)
plt.show()
```
