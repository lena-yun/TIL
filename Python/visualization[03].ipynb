```python
import folium as g
```

```python
# 37.509641, 127.055513
g_map = g.Map(location=[37.509641, 127.055513],
              zoom_start=18,
             tiles='Stamen Toner')
            #tiles : 지도의 스타일  (Stamen Terrain, Stamen Toner,,,)
#g_map

# option
marker01 = g.Marker([37.509641, 127.055513],
                 popup = 'campus seven',
                 icon  = g.Icon(color='blue'))
marker01.add_to(g_map)

marker02 = g.CircleMarker([37.509641, 127.055513],
                         radius=100,
                         color='skyblue',
                         popup='campus seven',
                         fill_color = 'blue') 
marker02.add_to(g_map)

g_map
#g_map.save('./data/now_location.map.html')
```

### elem_school_folium.xlsx

```python
xls = pd.ExcelFile('./data/elem_school_folium.xlsx')
seoul_school_df = xls.parse(xls.sheet_names[0])
seoul_school_df.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명주소, 위도, 경도
feature_sel_df = seoul_school_df.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
feature_sel_df.head() 
```

```python
## 소재지도로명주소가 서울을 포함하고 학교급구분이 초등학교인 것만 추출
# sql like
# 조건에 부합하는 결과를 데이터프레임으로 만들기 위해서 query()


#feature_sel_df['소재지도로명주소'].str.contains('서울')

#feature_sel_df['학교급구분'] =='초등학교' 
#feature_sel_df[feature_sel_df['학교급구분'] =='초등학교'].head()

# 결측값 확인,제거
print(feature_sel_df.isnull().sum())
feature_sel_df.dropna(inplace=True)
print('*'*50)
print(feature_sel_df.isnull().sum())
```

```python
result_df = feature_sel_df[ (feature_sel_df['학교급구분']=='초등학교') & (feature_sel_df['소재지도로명주소'].str[:5]=='서울특별시') ]
result_df.head()
```

```python
# 조건에 부합하는 결과를 데이터프레임으로 만들기 위해서 query()
#condition_df = feature_sel_df.query( "학교급구분 =='초등학교' and 소재지도로명주소.str.contains('서울특별시')" )

# 인덱스 추출
index_seoul = feature_sel_df['소재지도로명주소'].str.contains('서울')
#type(index_seoul)
index_ele = feature_sel_df['학교급구분']=='초등학교'
print(len(index_seoul))
print(len(index_ele))

result_df = feature_sel_df.loc[index_seoul & index_ele]
result_df.head()
```

```python
seoul_map = g.Map(location=[37.55, 126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

# 지도 시각화
for name,lat,lon in zip(result_df.학교명, result_df.위도, result_df.경도):
    g.Marker([lat, lon], 
             popup=name, 
             icon=g.Icon(color='purple')).add_to(seoul_map)
seoul_map
```

```python
pop_df = pd.read_excel('./data/경기도인구데이터.xlsx', index_col='구분')
pop_df.head()
#index_col='구분'  //'구분' column을 index로 설정
```

```python
# 경계정보를 가진 json 파일을 읽어보자
geo_data = json.load( open('./data/경기도행정구역경계.json', encoding='utf-8') )
geo_data
```

```python
pop_df.columns = pop_df.columns.map(str)
# 지도 생성
gyeonggi_map = g.Map(location=[37.509641, 127.055513],
                    zoom_start=10,
                    tiles='Stamen Terrain')

# g.Choropleth() : 구분된 경계를 표시할 때 사용하는 객체
year = '2017'
g.Choropleth(geo_data = geo_data,
            data = pop_df[year],
            columns = [pop_df.index, pop_df[year]],
            key_on='feature.properties.name').add_to(gyeonggi_map)
gyeonggi_map
```
